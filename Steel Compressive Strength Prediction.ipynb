## 1. Load Library and Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop

print(tf.__version__)


df = pd.read_csv('concrete.csv')
df.head()
```

## 2. Pre-Processing Step

```python
print(df.info())
print(df.isna().sum())
df.drop_duplicates(inplace=True)
print("Data yang ada duplikat: ", df.duplicated().sum())
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
n_comp = len(df.iloc[1, :-1].values)
print(n_comp)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

## 3. Build and Train The Model

```python
model = Sequential([
    Dense(64, input_shape=[n_comp], activation='relu'), 
    Dense(32, activation='relu'), 
    Dense(16, activation='relu'),
    Dense(1)
    ])
model.compile(optimizer=RMSprop(learning_rate=0.003), 
              loss='mean_squared_error', 
              metrics=['mean_absolute_error'])
history = model.fit(X_train, y_train, epochs=100, batch_size=15, validation_split=0.2)
loss, mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")
print(f"Test MAE: {mae}")
```

## 4. Evaluate Model Performance

```python
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
mse = mean_squared_error(y_true=y_test,y_pred=y_pred)
rmse = mean_squared_error(y_true=y_test,y_pred=y_pred,squared=False)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
y = np.array(y)
avg = np.mean(y)
relativeerror = mae/avg
relativeerrorpercentage = relativeerror * 100
print(f"Relative error model ini adalah {relativeerrorpercentage:.2f}%")
```
